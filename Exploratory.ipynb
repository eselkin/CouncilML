```python
from gensim.models import TfidfModel
from gensim.corpora import Dictionary
from gensim.models.phrases import Phrases, Phraser
from sklearn.cluster import DBSCAN
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}', '$', "“", "%", '#', '-', '_', '·', u'·', u'“', '/', '’'])
stop_words.update(list("abcdefghijklmnopqrstuvwxyz"))
import pandas as pd
from gensim.models import FastText
import numpy as np
from scipy.spatial.distance import cosine
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def is_float(string_to_test):
    assert(isinstance(string_to_test, str))
    try:
        float(string_to_test)
        return True
    except:
        return False
def words_in_sentence(sentence):
    acc = 0
    for word in sentence:
        acc += word[1]
    return acc
```

```python
# df here is the dataframe of the body corpus from 
df = pd.read_csv("./select_body_from_public_ingest_agendaite.csv")
df.columns = ['Body']
df = df.dropna()
# Vectorized string operations
# Remove {} from SQL array
df = df.apply(lambda x: x.str[1:-1])
# Split out the CSVness internal to a SQL array
df = df.apply(lambda x: x.str.split(","))
# Join list from column
df = df.apply(lambda x: x.str.join(" "))
```

```python
sentences = df["Body"].tolist() # pandas.Series -> list()
sentences_as_lists = [word_tokenize(sentence) for sentence in sentences]
sentences_as_lists = [[word.lower() for word in sentence] for sentence in sentences_as_lists]
```

```python
sentences_as_lists_without_stopwords = [[word for word in sentence if word not in stop_words and not is_float(word)] for sentence in sentences_as_lists]
sentences_without_stopwords = [" ".join(sentence) for sentence in sentences_as_lists_without_stopwords]
```

```python
# Bigramming corpus
phrases = Phrases(sentences_as_lists_without_stopwords, delimiter=b'_', min_count=1, threshold=1)
bigram = Phraser(phrases)
transformed_corpus = bigram[sentences_as_lists_without_stopwords]
```

```python
# Change format from TransformedCorpus to 2d list
sentences_with_bigrams = []
for doc in transformed_corpus:
    sentences_with_bigrams.append([word for word in doc])
```

```python
# Create a dictionary from the bigrammed documents
dictionary = Dictionary(sentences_with_bigrams)
```

```python
# Convert bigrammed documents into bag of words (id, freq)
bow_corpus = [dictionary.doc2bow(sentence) for sentence in sentences_with_bigrams]
```

```python
# Test to see how many words in a typical agenda item
print(words_in_sentence(bow_corpus[10]))
```

```python
# Create the TF-IDF model using the dictionary for terms
model = TfidfModel(bow_corpus, dictionary=dictionary)
print(model)
```

```python
# Generate the FastText unsupervised vectors from the corpus of the agenda items with bigrams
# Takes a bit of time to train
ft_model = FastText(sentences_with_bigrams, size=100, window=5, min_count=1, workers=4,sg=1, iter=20)
```

```python
# Get the TF-IDF scores of all words in all agenda items that have been bigrammed and put in BoW format
agitems_tfidf = []
for agitem_bow in bow_corpus:
    agitems_tfidf.append(model[agitem_bow])
```

```python
words_nparray = np.asarray(vectors)
print(words_nparray.shape)
```

```python
db = DBSCAN(eps=0.16, min_samples=27, metric='cosine').fit(words_nparray)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
```

```python
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]
        continue # It's mostly noise, just show non-noise clusters
    class_member_mask = (labels == k)

    xy = words_nparray[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = words_nparray[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```
